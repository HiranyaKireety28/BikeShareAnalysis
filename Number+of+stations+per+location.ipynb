```python
import pandas
import numpy as np
import matplotlib.pyplot as plt

df = pandas.read_csv('C:\\Users\\hiranya kireety\\Downloads\\bikeshareanalysis\\201402_station_data.csv',usecols = ["landmark","station_id"])
stnprcty = df.groupby('landmark',as_index = False).count()
fig,ax = plt.subplots()
ax.bar(np.arange(len(stnprcty)),stnprcty['station_id'], 0.6, color = "skyblue")
ax.set_xlabel('Cities')
ax.set_xticks(np.arange(len(stnprcty)))
ax.set_yticks(stnprcty['station_id'])
ax.set_xticklabels(stnprcty['landmark'])
ax.set_ylabel('Number of stations')
ax.set_title('Stations per City')
ax.set_yticklabels(stnprcty['station_id'])
plt.show()
```

```python
import matplotlib.pyplot as pt
df = pandas.read_csv('C:\\Users\\hiranya kireety\\Downloads\\bikeshareanalysis\\201402_station_data.csv',usecols = ['name','landmark'])
stnmsprcty = df.groupby('landmark')
cities = df['landmark'].drop_duplicates()
for city in cities:
    names = []
    names = stnmsprcty.get_group(city)
    fig,ax = pt.subplots()
    ax.set_title('Stations in '+city+' city')
    ax.set_xlabel('Station names')
    ax.legend(names['name'],loc = "best")
    percentage = [1 for i in list(numpy.arange(names['name'].count()))]
    ax.pie(percentage,labels = names['name'])
    pt.show()
```
