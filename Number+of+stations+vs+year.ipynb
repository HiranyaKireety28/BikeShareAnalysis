```python
import pandas
import matplotlib.pyplot as pt
import numpy
df = pandas.read_csv('C:\\Users\\hiranya kireety\\Downloads\\bikeshareanalysis\\201402_station_data.csv')
stations = df['station_id'].drop_duplicates() #stations list variable contains id's of all stations
pie_inp = [1 for i in numpy.arange(len(stations))]#assigning same value for all stations as they need to have equal width
pt.figure(figsize = (18,18))
pt.subplot(111)
pt.pie(pie_inp,labels = df['station_id'])#assigning labels depicting count of stations
pt.title(str(len(stations))+' Stations in 2014, First Half', color = "green")
pt.xlabel('Station ID\'s',color = 'brown')
pt.legend(df['name'],loc="best")
pt.show()
```
