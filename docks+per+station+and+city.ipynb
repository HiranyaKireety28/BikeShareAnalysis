```python
import pandas
import numpy
import matplotlib.pyplot as pt

df = pandas.read_csv('C:\\Users\\hiranya kireety\\Downloads\\bikeshareanalysis\\201402_station_data.csv',usecols = ['name','landmark','dockcount'])
fig_size = pt.rcParams["figure.figsize"]
fig_size[0] = 15
fig_size[1] = 15
pt.rcParams["figure.figsize"] = fig_size
fig,ax = pt.subplots()
ax.set_title('Dock per station')
ax.set_xlabel('Station')
ax.set_ylabel('Number of Docks')
ax.set_yticks(df['dockcount'])
ax.set_xticks(numpy.arange(df['name'].count()))
ax.set_xticklabels(df['name'],rotation = 90)
ax.set_yticklabels(df['dockcount'])
ax.bar(numpy.arange(df['name'].count()),df['dockcount'],0.7,color = "Pink")
pt.show()
```

```python
import pandas
import numpy
import matplotlib.pyplot as pt

df= pandas.read_csv('C:\\Users\\hiranya kireety\\Downloads\\bikeshareanalysis\\201402_station_data.csv',usecols = ['landmark','dockcount'])
docksprcity = df.groupby('landmark',as_index = False).count()
cities = df['landmark'].drop_duplicates()
fig,ax = pt.subplots()
ax.pie(docksprcity['dockcount'],labels = docksprcity['landmark'],autopct='%1.0f%%')
ax.set_title('Docks share per city')
ax.set_xlabel('Cities')
pt.show()

    
```
